```python
# 5; 1; 0

error_list = []
time_list = []

for i in time_output.iloc[0][1:-3]:
    error_list.append(time_output.iloc[1][10])
    time_list.append(time_output.iloc[3][10])
    
timesteps = time_output.iloc[0][1:-3].reset_index(drop = True)
diff_error = time_output.iloc[1][1:-3].reset_index(drop = True)
diff_time = time_output.iloc[3][1:-3].reset_index(drop = True)

fig, ax1 = plt.subplots()


color = 'tab:red'
ax1.set_xlabel('Number of timesteps per mesh movement')
ax1.set_ylabel('Mesh error', color=color)
#ax1.semilogx(diff_time.columns.values[2:-1], diff_time.iloc[0][2:-1], ':', label = r'$\alpha = 3$', color=color)
ax1.semilogx(timesteps, diff_error, '-o', label = 'Mesh movement', color=color)
#ax1.semilogx(diff_time.columns.values[2:-1], diff_time.iloc[2][2:-1], ':', label = r'$\alpha = 5$', color=color)
ax1.semilogx(timesteps, error_list, '--', label = 'Fixed mesh', color = color)
ax1.tick_params(axis='y', labelcolor=color)
plt.legend(loc = 9, bbox_to_anchor=(0.5, 0.73), title = "Error")
ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('Computational time (s)', color=color)  # we already handled the x-label with ax1
#ax2.semilogx(diff_time.columns.values[2:-1], diff_time.iloc[8][2:-1], ':', label = r'$\alpha = 3$', color=color)
ax2.semilogx(timesteps, diff_time, '-o', label = 'Mesh movement', color=color)
#ax2.semilogx(diff_time.columns.values[2:-1], diff_time.iloc[10][2:-1], ':', label = r'$\alpha = 5$', color=color)
ax2.semilogx(timesteps, time_list, '--', label = 'Fixed mesh',  color = color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.legend(loc = 9, bbox_to_anchor=(0.5, 0.95), title = "Cost")
plt.show()
```
