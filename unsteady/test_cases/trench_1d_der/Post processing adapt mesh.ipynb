```python
results_fixed = pd.read_excel('results.xls', sheet_name = 'fixed', index_col = 0)
```

```python
from matplotlib.ticker import ScalarFormatter, FormatStrFormatter


fixed_mesh = results_fixed.loc['fixed disc'][:6]
fixed_exp = results_fixed.loc['fixed exp'][:6]

min_mesh = results_fixed.loc['min disc'][:6]
min_exp = results_fixed.loc['min exp'][:6]

gen_mesh = results_fixed.loc['gen disc'][:6]
gen_exp = results_fixed.loc['gen exp'][:6]

no_elem = results_fixed.loc['no elem'][:6]

fig, ax = plt.subplots()
ax.loglog(no_elem, fixed_mesh, '*-', label = 'Fixed mesh')
ax.loglog(no_elem, gen_mesh, '*-', label = 'General parameters')
ax.loglog(no_elem, min_mesh, '*--', label = 'Optimum parameters')
ax.set_xticks([10, 20, 30, 40, 60])
ax.set_xticklabels([10, 20, 30, 40, 60])
ax.set_xlabel(r'Number of mesh elements in $x$-direction')
ax.set_ylabel("Pointwise L2 error norm")
plt.grid()
plt.legend()
plt.show()

fig1, ax1 = plt.subplots()
ax1.plot(no_elem, fixed_exp, '*-', label = 'Fixed mesh')
ax1.plot(no_elem, gen_exp, '*-', label = 'General parameters')
ax1.plot(no_elem, min_exp, '*--', label = 'Optimum parameters')
ax1.set_xticks([10, 20, 30, 40, 60])
ax1.set_xticklabels([10, 20, 30, 40, 60])
ax1.set_xlabel(r'Number of mesh elements in $x$-direction')
ax1.set_ylabel("Pointwise L2 error norm")
plt.legend()
plt.grid()
plt.show()
```
