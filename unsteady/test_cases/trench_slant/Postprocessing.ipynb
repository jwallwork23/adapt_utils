```python
diff_time = pd.read_excel('/Users/mc4117/Documents/adapt_utils/unsteady/test_cases/trench_slant/results.xlsx', sheet_name = 'Sheet3')

error_list = []
time_list = []

for i in diff_time.columns.values[2:-2]:
    error_list.append(diff_time.iloc[0][0])
    time_list.append(diff_time.iloc[1][0])

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Number of timesteps per mesh movement')
ax1.set_ylabel('Mesh error', color=color)
#ax1.semilogx(diff_time.columns.values[2:-1], diff_time.iloc[0][2:-1], ':', label = r'$\alpha = 3$', color=color)
ax1.loglog(diff_time.columns.values[2:-2], diff_time.iloc[0].reset_index(drop = True)[2:-2], '-o', label = 'Mesh movement', color=color)
#ax1.semilogx(diff_time.columns.values[2:-1], diff_time.iloc[2][2:-1], ':', label = r'$\alpha = 5$', color=color)
ax1.loglog(diff_time.columns.values[2:-2], error_list, '--', label = 'Fixed mesh', color = color)
ax1.tick_params(axis='y', labelcolor=color)
plt.legend(loc = 9, bbox_to_anchor=(0.5, 0.73), title = "Error")
ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('Computational time (s)', color=color)  # we already handled the x-label with ax1
#ax2.semilogx(diff_time.columns.values[2:-1], diff_time.iloc[8][2:-1], ':', label = r'$\alpha = 3$', color=color)
ax2.loglog(diff_time.columns.values[2:-2], diff_time.iloc[1].reset_index(drop = True)[2:-2], '-o', label = 'Mesh movement', color=color)
#ax2.semilogx(diff_time.columns.values[2:-1], diff_time.iloc[10][2:-1], ':', label = r'$\alpha = 5$', color=color)
ax2.loglog(diff_time.columns.values[2:-2], time_list, '--', label = 'Fixed mesh',  color = color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.legend(loc = 9, bbox_to_anchor=(0.5, 0.95), title = "Cost")
plt.show()
```
