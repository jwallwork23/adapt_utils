```python
import pandas as pd
import pylab as plt
```

```python
tsunami_opt_fix = pd.read_excel('tsunami_bump.xls', sheet_name = 'Sheet1', index_col =0)
tsunami_nx1 = pd.read_excel('tsunami_bump.xls', sheet_name = 'nx 1 ny 1')
tsunami_nx2 = pd.read_excel('tsunami_bump.xls', sheet_name = 'nx2 ny2')
tsunami_nx3 = pd.read_excel('tsunami_bump.xls', sheet_name = 'nx3 ny3')
tsunami_nx4 = pd.read_excel('tsunami_bump.xls', sheet_name = 'nx4 ny 4')
```

```python
import numpy as np
logx = [np.log(i) for i in tsunami_opt_fix.loc['no_elem'][1:-1].values]
log_error = [np.log(i) for i in tsunami_opt_fix.loc['fixed_err'][1:-1].values]
poly = np.polyfit(logx, log_error, 1)
print(poly[0])

y = [-x + poly[1] for x in logx]
```

```python
fig, ax = plt.subplots()
ax.plot(tsunami_opt_fix.loc['no_elem'][1:], tsunami_opt_fix.loc['fixed_err'][1:].values, '*-')
ax.plot([np.exp(i) for i in logx], [np.exp(i) for i in y], 'k--', label = '1st order')
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel(r'Number of mesh elements in $x$-direction')
ax.set_ylabel(r'L2 error norm')
ax.set_xticks([60, 90, 120, 150])
ax.set_xticklabels([60, 90, 120, 150])
ax.set_yticks([0.075, 0.1, 0.15, 0.2])
ax.set_yticklabels([0.075, 0.1, 0.15, 0.2])
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.semilogy(tsunami_opt_fix.loc['fixed_time'][1:], tsunami_opt_fix.loc['fixed_err'][1:].values, '*-', label = 'Fixed mesh')
ax.semilogy(tsunami_opt_fix.loc['gen_time'], tsunami_opt_fix.loc['gen_err'].values, 'k*', ms =8, label = 'General parameters')
ax.semilogy(tsunami_opt_fix.loc['opt_time'], tsunami_opt_fix.loc['opt_err'].values,  '^', ms = 5, label = 'Optimum parameters')
plt.legend(fontsize = 11.5)
ax.set_xticks([0, 1000, 2000, 3000, 4000, 5000])
ax.set_xticklabels([0, 1000, 2000, 3000, 4000, 5000])
ax.set_yticks([0.04, 0.06, 0.1, 0.2])
ax.set_yticklabels([0.04, 0.06, 0.1, 0.2])
ax.set_xlabel('Computational cost (s)')
ax.set_ylabel('L2 error norm')
plt.grid(which='major')
plt.show()
```

```python
tsunami_freq = pd.read_excel('tsunami_bump.xls', sheet_name = 'freq', header = None)[:4]
```

```python
error_list = []
time_list = []
fixed_error_list = []
fixed_time_list = []

for i in tsunami_freq.columns.values[1:]:
    error_list.append(tsunami_freq.iloc[2][i])
    time_list.append(tsunami_freq.iloc[3][i])
    fixed_error_list.append(tsunami_freq.iloc[2][0])
    fixed_time_list.append(tsunami_freq.iloc[3][0])  
```

```python
tsunami_freq.iloc[3]
```

```python
import matplotlib.ticker as mtick
fig, ax1 = plt.subplots()

plt.gca().ticklabel_format(style='sci', scilimits=(0,1), axis='y')

color = 'tab:red'
ax1.set_xlabel('Number of timesteps per mesh movement')
ax1.set_ylabel('Discretisation error', color=color)
ax1.semilogx(tsunami_freq.iloc[1][1:-2], [i/fixed_error_list[0] for i in error_list[:-2]], '-o', label = 'Mesh movement', color=color)
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_yticks([0.44, 0.45, 0.46])
ax1.set_yticklabels([0.44, 0.45, 0.46])
ax1.yaxis.set_major_formatter(mtick.FuncFormatter(lambda y, _: '{:.0%}'.format(y))) 
plt.minorticks_off()

legend = plt.legend(fontsize = 11, loc = 9, bbox_to_anchor=(0.32, 0.85), title = "Error")
legend.get_title().set_fontsize('11')
ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('Computational cost', color=color)  # we already handled the x-label with ax1
ax2.semilogx(tsunami_freq.iloc[1][1:-2], [i/fixed_time_list[0] for i in time_list[:-2]], '-o', label = 'Mesh movement', color=color)
ax2.tick_params(axis='y', labelcolor=color)
ax2.yaxis.set_major_formatter(mtick.FuncFormatter(lambda y, _: '{:.0%}'.format(y))) 


fig.tight_layout()  # otherwise the right y-label is slightly clipped
legend = plt.legend(fontsize = 11, loc = 9, bbox_to_anchor=(0.32, 1), title = "Cost")
legend.get_title().set_fontsize('11')
plt.show()
```
